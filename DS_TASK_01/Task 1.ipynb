## Business Problem Understanding

* Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.
* Here i done 3 projects for Task 1, two projects by creating own data from pandas and one (3rd) project from csv file.

### Importing Libraries and Dataset 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Generating the Data 

```python
# Generate a random dataset for ages in a population
np.random.seed(42)  # Setting seed for reproducibility
ages = np.random.randint(18, 65, size=100)  # Generate 100 random ages between 18 and 65
```

```python
# Create a DataFrame
df = pd.DataFrame({'Age': ages})
```

### Visualization 

```python
# Plotting a histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Ages in a Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 2nd Project Countries vs Population

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a fictional dataset of countries and their populations
np.random.seed(42)  # Setting seed for reproducibility
```

```python
countries = ['CountryA', 'CountryB', 'CountryC', 'CountryD', 'CountryE']
populations = np.random.randint(10000000, 500_000_000, size=len(countries))
```

```python
df = pd.DataFrame({'Country': countries, 'Population': populations})
```

```python
# Plotting a bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Country'], df['Population'], color='lightblue')
plt.title('Distribution of Populations in Fictional Countries')
plt.xlabel('Country')
plt.ylabel('Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Filter out rows with missing values in the '2019' column
df_filtered = df[['Country Name', '2019']].dropna()
```

```python
# Sort the data by the '2019' column in descending order
df_sorted = df_filtered.sort_values(by='2019', ascending=False)
```

```python
# Select the top 10 countries by population
top_10_countries = df_sorted.head(10)
```

```python
# Plotting a bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries['Country Name'], top_10_countries['2019'], color='blue')
plt.title('Top 10 Countries by Population in 2019')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')  # Rotate country names for better readability
#plt.tight_layout()
plt.show()
```
