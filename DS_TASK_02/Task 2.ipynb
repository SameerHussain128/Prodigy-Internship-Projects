# Data Understanding

### Importing Libraries and Dataset 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\Data Science 6pm\Projects\EDA\titanic\train.csv')
```

```python
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing Values:\n", missing_values)
```

#### There are missing values 

#  Data Cleaning

```python
# Fill missing values for 'Age' with the median
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Fill missing values for 'Embarked' with the mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# Drop 'Cabin' column due to high missing values
df.drop('Cabin', axis=1, inplace=True)
```

##  EDA - Exploratory Data Analysis
### Visualization through Graphs and charts

```python
# Visualize the survival distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=df, palette='Set1')
plt.title('Survival Distribution')
plt.show()
```

### Insights :
* Less num of peoples are survived

```python
# Explore the relationship between Pclass and Survival
plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set2')
plt.title('Survival by Pclass')
plt.show()
```

### Insights :
* More peoples died in Pclass 3

```python
# Explore the distribution of ages
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

### Insights :  
* More no of peoples on ship are age between 25-30

```python
# Explore the relationship between Sex and Survival
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set3')
plt.title('Survival by Sex')
plt.show()
```

### Insights :
* More no of mens are dies where as womens died very less
* Mens survival rate is very less compared to womens

```python
# Explore the relationship between Embarked and Survival
plt.figure(figsize=(8, 5))
sns.countplot(x='Embarked', hue='Survived', data=df, palette='Set1')
plt.title('Survival by Embarked')
plt.show()
```

### Insights :
* More no of peoples are died in Embarked S
* Very less no of peoples are died in Embarked C and D. Survival rate is also very less
